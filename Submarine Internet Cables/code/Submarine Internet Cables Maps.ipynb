```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the submarine cables dataset (assumes GeoJSON format)
cables = gpd.read_file('cables.geojson') # make sure to put the file in the same folder as the notebook

# Load a world map dataset provided by Geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a new column with colors for each country
world['colors'] = 'black'

# Update the color for Cote d'Ivoire to be orange
world.loc[world['name'] == "Ivory Coast", 'colors'] = 'orange'

# Plot the world map with a black background
ax = world.plot(figsize=(20, 15), linewidth=0.25, edgecolor='black', facecolor='black', alpha=0.5)

# Plot the submarine cables on top of the world map with a color map
cables.plot(ax=ax, linewidth=2, cmap='Paired', alpha=0.9)

# Customize the appearance of the map
ax.set_title('Submarine Internet Cables')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_axis_off()

# Set the background color of the plot to black
ax.set_facecolor('black')

# Display the map
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the submarine cables dataset (assumes GeoJSON format)
cables = gpd.read_file('cables.geojson') # make sure to put the file in the same folder as the notebook

# Create an empty plot with a black background
fig, ax = plt.subplots(figsize=(20, 12))
ax.set_facecolor('black')

# Plot the submarine cables with a color map
cables.plot(ax=ax, linewidth=1, cmap='viridis', alpha=0.7)

# Customize the appearance of the map
ax.set_title('Submarine Internet Cables', color='white')
ax.set_xlabel('Longitude', color='white')
ax.set_ylabel('Latitude', color='white')
ax.set_axis_off()

# Set the background color of the plot to black
plt.gca().set_facecolor('black')

# Display the map
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the submarine cables dataset (assumes GeoJSON format)
cables = gpd.read_file('cables.geojson')

# Create an empty plot with a black background
fig, ax = plt.subplots(figsize=(25, 15))
ax.set_facecolor('black')


# Plot the submarine cables on top of the world map with a color map
cables.plot(ax=ax, linewidth=2, cmap='Paired', alpha=0.8)


# Customize the appearance of the map
ax.set_title('Submarine Internet Cables', color='white')
ax.set_xlabel('Longitude', color='white')
ax.set_ylabel('Latitude', color='white')
ax.tick_params(axis='both', colors='white')  # Change the tick colors to white
ax.spines['bottom'].set_color('white')  # Change the axis color to white
ax.spines['left'].set_color('white')  # Change the axis color to white

# Display the map
plt.show()
```
